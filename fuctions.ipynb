```python
#Function's
def is_even(num):
    """
    this function returns given num is odd/even
    input - any valid integer
    output - even/odd
    created on - 11/Jan/2026
    """
    if type(num) == int:
    
        if num % 2 == 0:
            return "even"
        else:
            return "odd"
    else:
        return "Abe Chutiya hai kya..? integer daal"
```

```python
for i in range(1,11):
    x = is_even(i)
    print(x)
    
```

```python
is_even("Hello world")
```

```python
#types of argument
# 1> Default argument
# 2> Positional argument
# 3> Keyword argument

#Default argument
def power(a = 1, b = 1):
    return a**b
    
power(2)
```

```python
#positional argument
power(2,3)
```

```python
#keyword argument
power(b = 3, a = 2)
```

```python
#*args and **kwargs

def multiply(*args):
    product = 1
    for i in args:
        product = product * i
    return product
```

```python
multiply(1,2,3,4,5,78,5,3)
```

```python
#**kwargs
def display(**kwargs):
    for (key, value) in kwargs.items():
        print(key,'->', value)
display(india = "Delhi", Nepla = "Kathmandu")
```

```python
def squre(x):
    return x**2
def transform(f,l):
    output = []
    for i in l:
        output.append(f(i))
    print(output)

l = [1,2,3.4,5]
# transform(squre, l) #-> squre ki jagah lambda function use kar sakte hai
transform(lambda x: x**2, l)
```
