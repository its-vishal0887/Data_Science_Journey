```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#categorical plot
#strip plot

tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.stripplot(data=tips, x= 'day', y = 'total_bill', hue='day')
```

```python
sns.catplot(tips, x = 'day', y = 'total_bill', jitter = 0.2, hue='sex')
```

```python
#swarm plot
sns.catplot(tips, x = 'day', y = 'total_bill', kind='swarm')
```

```python
swarm = sns.swarmplot(tips, x = 'day', y = 'total_bill', hue = 'sex')
swarm
```

```python
#box plot
sns.boxplot(tips, x= 'day', y='total_bill', hue='day')
plt.title('Box plot')
plt.grid()
```

```python
sns.catplot(tips, x= 'day', y='total_bill', hue='sex', kind = 'box')
plt.title('Box plot')
plt.grid()
```

```python
sns.boxplot(tips, y='total_bill')
```

```python
#violin plot
sns.violinplot(tips, x = 'day', y = 'total_bill')
```

```python
sns.catplot(tips, x= 'day', y = 'total_bill', kind='violin', hue = 'sex', split= True)
```

```python
#categorical estimate plot
#bar plot
sns.barplot(tips, x='sex', y='total_bill', hue='smoker')
```

```python
#point plot
sns.pointplot(tips, x='day', y='total_bill', hue='sex')
```

```python
#count plot
sns.countplot(tips, x = 'sex', hue='smoker')
```

```python
#faceting using catplot
```

```python
sns.catplot(tips, x= 'sex', y = 'total_bill', col = 'smoker',row = 'time', kind='violin')
```

```python
#regplot
sns.regplot(tips, x = 'total_bill', y = 'tip')
```

```python
sns.lmplot(tips, x = 'total_bill', y='tip', hue = 'sex')
```

```python
#residual plot
sns.residplot(tips, x = 'tip', y = 'total_bill')
```
