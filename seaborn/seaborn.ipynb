```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Relational plot
snse = sns.scatterplot(tips, x= 'total_bill', y = 'tip', hue='sex')
snse
```

```python
#figure level plotting

sns.relplot(tips, x= 'total_bill', y = 'tip', hue='sex', style='time', size='size')
```

```python
#line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
sns.lineplot(temp_df, x='lifeExp', y='year')
```

```python
#fig level graph
sns.relplot(temp_df, x='year', y='lifeExp', kind='line')
```

```python
#figure level plot
df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
sns.relplot(df, x='year', y= 'lifeExp', kind='line', hue='country', style='continent')
```

```python

sns.relplot(tips, x= 'total_bill', y = 'tip',kind='line', col = 'sex', row= 'day')
```

```python
#distrubuted graphs
#1 histplot
sns.histplot(tips, x='total_bill')
```

```python
#using figure level attribute
sns.displot(tips, x= 'total_bill', kind = 'hist')
```

```python
#hue attribute
sns.displot(tips, x= 'total_bill', kind = 'hist', hue = 'sex')
```

```python
titanic = sns.load_dataset('titanic')
sns.displot(titanic, x= 'age', kind = 'hist', col = 'sex')
```

```python
sns.kdeplot(tips, x='total_bill')
sns.rugplot(tips, x = 'total_bill')
```
