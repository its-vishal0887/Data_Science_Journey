```python
import numpy as np
```

```python
#1 D array isko vector kehte hai
a = np.array([i for i in range(1,10)])
print(a)
```

```python
# 2D array isko metrix kehte hai
a = np.array([[j for j in range(1, 10)] for i in range(1,10) ])
print(a)
```

```python
#3 D array isko hi tensor array kehte hai
a = np.array([[[n for n in range(1,10)] for j in range(1,4)] for i in range(1,4)])
print(a)
```

```python
# changing data type of array
a = np.array([i for i in range(1,10)], dtype=float)
b = np.array([i for i in range(1,10)], dtype=str)
c = np.array([i for i in range(1,10)], dtype=bool)
d = np.array([i for i in range(1,10)], dtype=complex)
print(a)
print(b)
print(c) #non zero element ko python True ki tarah treat karta hai
print(d)
```

```python
#Array Operation's
a2 = np.arange(12).reshape(3,4)
a1 = np.arange(12, 24).reshape(3,4)
```

```python
#SCALER OPERATION
#arithmatic operation's
print("->", a1 + 1)
print("->", a1 - 1)
print("->", a1 * 1)
print("->", a1 ** 2)
print("->", a1 / 2)
print("->", a1 % 2)
```

```python
#Relational Operation's
print("->", a1 > 5) #iska matlab hai each element from a1 is greater the 5 or not
print("->", a1 ==1)
print("->", a1 < 5)
print("->", a1 != 0)
print("->", a1 >= 5)
print("->", a1 <= 5)
```

```python
#Vector Operation's

a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
print("a1 ->  ", a1)
print("a2 ->  ", a2)
```

```python
print("->", a1 > a2)
print("->", a1 < a2)
print("->",a1 >= a2)
print("->", a1 <= a2)
print("->", a1 == a2)
print("->", a1 != a2)
```

```python
#Array Function's

a1 = np.random.random((3,3))
print('A1 -> ', a1)
a1 = np.round(a1 * 100) #decimal ko nearest element me convert karta hai
print('nearest ->', a1)
```

```python
#mathmatical function of array
#max / min / sum / prod / axis = 0 -> col and axit = 1 -> row's
print(np.max(a1, axis=0)) #max -> ye batata hai ki pure array me se max element kya hai
print(np.min(a1 , axis = 1)) #min element batata hai
print(np.sum(a1, axis = 0)) #pure array ka sum calculate karega
print(np.prod(a1, axis = 1)) #pure array ka product nika ke deta hai
```

```python
#some statistical operation's
#mean / median / standard deviation(std) / veriance(var)
a1 = np.arange(12).reshape(3,4)
a1
```

```python
print(np.mean(a1, axis = 1))
print(np.median(a1, axis = 0))
print(np.std(a1, axis= 1))
print(np.var(a1, axis = 0))
```
