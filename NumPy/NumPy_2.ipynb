```python
import numpy as np
#sigmoid -> ye ek math ka function hai jo kisi bhi num ko 0 - 1 ke beech me laa deta hai

""" return 1 / (1 + np.exp(-x))"""

def sigmoid(array):
  array = np.array(array)
  return 1 / (1 + np.exp(-array))

print(sigmoid([10, 5, 1]))
```

```python
#mean squared error
#ye ek ML ka loss function hai
actural = np.random.randint(1, 50, 25)
predicted  =np.random.randint(1 , 50, 25)
MSE = np.mean(((predicted - actural)**2))
print(MSE)
```

```python
#Categorial class entropy
#ye bhi ML ka loss function hai

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1, 5,25)

CCE = -np.sum(actual * np.log(predicted))
print(CCE)
```

```python
#Binary Cross Entropy

# For BCE, y should be true labels (0 or 1) and p should be predicted probabilities (0 to 1)
y = np.random.randint(0, 2, 25) # True labels (0 or 1)
p = np.random.rand(25)       # Predicted probabilities (0 to 1)

BCE = -(y*np.log(p) + (1-y)* np.log(1-p))

print(BCE)
```
