```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/iris.csv')
df['Species'] = df['Species'].replace({'Iris-setosa' : 0, 'Iris-virginica':1, 'Iris-versicolor':2})
df.sample(5)
```

```python
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c=df['Species'], alpha=0.7, edgecolors='black')
plt.xlabel("Sepal length")
plt.ylabel("petal length")
plt.colorbar()
plt.title("Iris")
```

```python
#anotation on graph
plt.figure(figsize=(15,10))
df1 = pd.read_csv('/content/batter.csv').head(25)
plt.scatter(df1['avg'], df1['strike_rate'], s= df1['runs'],color = 'skyblue', alpha=0.7, edgecolors='black')
plt.axhline(130,color = 'black' )
plt.axvline(40, color = 'black')
for i in range(df1.shape[0]):
  plt.text(df1['avg'].values[i], df1['strike_rate'].values[i], df1['batter'].values[i])
```

```python
fig , ax = plt.subplots(figsize = (10,8))
ax.scatter(df1['avg'], df1['runs'] )
ax.set_xlabel("avg")
ax.set_ylabel("runs")
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (12,8), sharex = True)

ax[0].scatter(df1['avg'], df1['strike_rate'])
ax[1].bar(df1['avg'], df1['runs'])

ax[1].set_title("runs")
ax[0].set_title('strike rate')

ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike rate')

ax[1].set_xlabel('batter')
ax[1].set_ylabel('runs')


fig.show()
```

```python
from matplotlib import markers
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (12,8))

ax[0][0].scatter(df1['avg'], df1['strike_rate'])
ax[0][1].bar(df1['avg'], df1['runs'])

ax[0][1].set_title("runs")
ax[0][0].set_title('strike rate')

ax[0][0].set_xlabel('avg')
ax[0][0].set_ylabel('strike rate')

ax[0][1].set_xlabel('batter')
ax[0][1].set_ylabel('runs')

ax[1][0].plot(df1['batter'], df1['strike_rate'], marker = '^')

ax[1][1].hist(df1['strike_rate'])



fig.show()
```

```python
#3d graph's

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter3D(df1['avg'], df1['runs'], df1['strike_rate'], marker = '^')
ax.plot3D(df1['avg'], df1['runs'], df1['strike_rate'])
ax.set_title("Bat's man runs analysis")
ax.set_xlabel("Avg")
ax.set_ylabel("Runs")
ax.set_zlabel("Strike rate")
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2


fig = plt.figure(figsize=(18,10))
```

```python
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy, z, cmap = 'viridis')
```

```python


x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = np.sin(xx) + np.cos(yy)


fig = plt.figure()
ax = plt.subplot()
ax.contourf(xx,yy, z, cmap = 'viridis')
```
