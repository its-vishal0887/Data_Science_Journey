```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman.head(1)
```

```python
#graph ploting
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
#labeling
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python

#changing color's
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green')

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python

#changing line style
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black', linestyle = 'dotted')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green', linestyle = 'dashdot')

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# changing width of lines

#changing line style
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black', linestyle = 'dotted', linewidth = 2)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green', linestyle = 'dashdot', linewidth = 3)

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# adding markers
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black', linestyle = 'dotted', linewidth = 2, marker = '>')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green', linestyle = 'dashdot', linewidth = 3, marker = '>')

plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend() -> label
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black', linestyle = 'dotted', linewidth = 2, marker = '>', label = 'Rohit')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green', linestyle = 'dashdot', linewidth = 3, marker = '>', label = 'Virat')

plt.legend()
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# limiting axes   ylim(range)
price = [40000, 54000,57000,49000,47000,450000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year, price)
plt.ylim(0, 100000)
```

```python
# limiting x axis
price = [40000, 54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year, price)
plt.xlim(2016,2018)
```

```python
# Grids
plt.plot(batsman['index'], batsman['V Kohli'], color = 'black', linestyle = 'dotted', linewidth = 2, marker = '>', label = 'Rohit')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'green', linestyle = 'dashdot', linewidth = 3, marker = '>', label = 'Virat')

plt.legend()
plt.title('Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
```
