Patterns
-> Nested Loops

```python
n = int(input("Enter N: "))

for i in range(1, n+1):
  for j in range(1, i):
    print("*", end = " ")
  print()
```

```python
n = int(input("Enter N: "))

for i in range(1, n+1):
  for j in range(1, n-i+1):
    print(" ", end = " ")
  for k in range(n-i, n):
    print("*", end = " ")
  print()
```

```python
n = int(input("Enter N: "))
for i in range(1, n+1):
  for j in range(1, i+1):
    print(" ", end= " ")
  for k in range(i, n+1):
    print("*", end = " ")
  print()
```

```python
n = int(input("Enter N: "))

for i in range(1, n+1):
  for j in range(1, i+1):
    print(j, end = " ")
  for j in range(i-1, 0, -1):
    print(j, end= " ")
  print()
```

```python
lower = int(input("Entr Lower Range: "))
upper = int(input("Enter Upper Range: "))

for i in range(lower, upper+1):
  for j in range(2, i):
    if i % j == 0:
      break
  else:
    print(i)

```

Common Function in Python(Apply on all type of data types)

```python
#1) len()
len("Hello World") #retun length of string as 11

#2) max()
max("Hello World") #return max ASCII value character in string

#3) min()
min("Hello World") #return min ASCII value character in string

#4) sorted()
sorted("Hello World") #return list of character in sroted ascending format base on ASCII values
sorted("Hello World", reverse=True) #in sorted descending format
```

Functions Only For String

```python
#1) capitalized
s = "hello world"
s.capitalize() #capitalize starting character of string

#2) title()
s = "hello world"
s.title() #capitalized each sarting character of sub string

#3) upper()
s = "hello world"
s.upper() #convert each character of string in uppercase

#4) lower()
s = 'HELLO WORLD'
s.lower() #convert each character of string in lowercase

#5) swapcase()
s = 'Hello World'
s.swapcase() #convert lower character into upper and vice versa
```

Other Function's 
1) count() 2) find() 3) index()

```python
#count()
s = "my name is vishal"
s.count("i") #return occurance of i and any substring

#find()
s = "my name is Vishal"
s.find("is") #return first occurace in string 

#index()
s = "my name is Vishal"
s.index("is") #same as find
```

```python
#1) endswith()
s = "my Name is Vishal"
s.endswith('is') #if string is end with is then it will return true

#2) startswith()
s = "my name is Vishal"
s.startswith('my') #return true if string start with my

#3) format()

s = "my name is {} and i am {}"
s.format('Vishal', 'Male')
```

```python
#isalnum()
#return true if string contain's only alphabate's and numbers
s = 'Vishal123'
s.isalnum()

#isaplha()
s = 'Vishal@12'
s.isalpha() #return true if and only if string contains alphabate's

#isdigit()
#return's true if string contain's only digit's
s = '1234'
s.isdigit()

#isidentifier()
s = "Vishal123"
s.isidentifier()
#return true if string is in identifier's form
```

#Most IMP function's 
1) split()
2) join()

```python
#1) split()
s = "my name is vishal"
s.split() # split string in to list of sub-string's ['my', 'name', 'is', 'vishal']
s.split('i') # this will break i from string and from list of substring's

#2) join()

" ".join(['my', 'name', 'is', 'vishal'])

#this will join list of string into proper string -> 'my name is vishal'

'-'.join(['my', 'name', 'is', 'vishal']) #-> 'my-name-is-vishal'

```

```python
#replace()
s = "hi my name is vishal"
s.replace('vishal', 'dhiresh') #-> 'hi my name is dhiresh'

#strip()

s = ' hi '
s.strip() #-> 'hi' it removes leading and trailing spaces
```
