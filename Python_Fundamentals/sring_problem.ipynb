```python
n = input("Enter String: ")
i = 0
j = len(n)-1

while i <= j :
    if n[i] == n[j] :
        i = i+1
        j = j - 1
    else:
        break


if i >= j :
    print("String Is Plaindrome..!")
else:
    print("Stirng is Not Palindrome..!")
```
