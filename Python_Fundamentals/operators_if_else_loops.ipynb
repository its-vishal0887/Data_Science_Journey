Program - Find the sum of a 3 digit number entered by the user.

```python
num = int(input("Enter 3 Digit Num: "))

sum = 0
while(num != 0):
  newNum = num % 10
  num = num//10
  sum = sum + newNum

print(sum) 
```

Find Min of 3 Num

```python
num_1 = int(input("Enter Num_1: " ))
num_2 = int(input("Enter Num_2: "))
num_3 = int(input("Enter Num_3: "))

if(num_1 < num_2 and num_1 < num_3):
  print(f"{num_1} is Smallest!!")
elif(num_2 < num_1 and num_2 < num_3):
  print(f"{num_2} is Smallest!!!")
else:
  print(f"{num_3} is Smallest!!..")

```

Menu Deriven Calculator

```python
fnum = int(input("Enter F_num: "))
snum = int(input("Enter S_num: "))

op = input("Enter Which Operation You Want To Perform?(just enter +,-,*,/,//,%): ")
if(op == '+'):
  print(f"Addition: {fnum + snum}")
elif(op == '-'):
  print(f"SubStraction: {fnum - snum}")
elif(op == '*'):
  print(f"Multiplication: {fnum * snum}")
elif(op == '/'):
  print(f"Division: {fnum/snum}")
elif(op == '//'):
  print(f"Integer Division: {fnum//snum}")
elif(op == '%'):
  print(f"Modulo Opertion: {fnum % snum}")
else:
  print("Enter Valid Operation!!..")
```
