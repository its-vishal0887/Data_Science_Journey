```python
#empty
print([])
#1D
print([1,2,3,4,5])
#2D
print([1,2,3,4,[1,2,3]])
#3D
print([1,5,6,[4,5,6,[7,9,6]]])
#type Conversion
print(list("Hello"))
```

```python
#indexing 
l = [1,2,3,4,[7,8]]
#positive
print(l[0])
#negative
print(l[-1])

#for 2D
print(l[4])

#for 3D

s = [[1,2],[1,5,[7,9]]]
print(s[1][2][0])
```

```python
#Slicing
l = [1,2,3,4,5,6]
print(l[0:4:2])
```

```python
#insertion
#append() try to adding only one element at a time in ending of list
l = [1,2,3,4,5,6]
l.append([1,2,3])
print(l)
```

```python
#extend() trying to add multiple item into the list at a time
l = [1,2,3,4,5,6]
l.extend("Delhi")
print(l)
```

```python
#insert(index,item)
l = [1,2,3,4,5,6]
l.insert(0,100)
print(l)
```

```python
#editing item in list
#indexing
l = [1,2,3,4]
l[-1] = 500;
print(l)

#Slicing
l = [1,2,3,4,5]
l[1:4] = [200,300,400]
print(l)
```

```python
#Deletion
#1 dell
l = [1,2,3,4,5,6]
#indexing
del l[0] #deleting specific index element
print(l)

#slicing
del l[0:3]
print(l)
```

```python
#remove
l = [1,2,3,4,5,6]
l.remove(5) #remove specific element from list
print(l)
```

```python
#3 pop()
l = [1,2,3,4,5,6]
l.pop(0) #remove from index
print(l)

l.pop() #by default remove laste index element
print(l)
```

```python
#clear()-> return empty list
l = [1,2,3,4,5,6]
l.clear()
print(l)
```

```python
a = [1,2,3]
b = a.copy()
print(b)

a.append(4)
print(b)
```

```python
#Operation on list
#1> Arithmatic
#2> Membership
#3> loop

#1 Arithmatic
#1 (+)
l1 = [1,2,3,4]
l2 = [5,6,7,8]
print(l1 + l2)

#2 (*)
print(l1 * 3)
```

```python
#membarship
#1 in
l1 = [1,2,3,4]
print(4 in l1)

#2 not in
print(6 not in l1)
```

```python
#loop
l1 = [1,2,3,4]
for i in l1:
    print(i)
```

```python
#list function's
#1 len()
l1 = [1,2,3,4]
print(len(l1))

#2 min()

print(min(l1))

#3 max()
print(max(l1))

# sorted()
print(sorted(l1,reverse = True)) #temporary sort

#sort()-> parmanently sort
l1.sort()
print(l1)

#copy()

l2 = l1.copy()
print(l2)

```

```python
#list comprehnsion
#add 1 to 10 number to a list
l = [i for i in range(1,11)]
print(l)
```

```python
#scalar multiplication on a vector
v = [2,3,4]
s = -3
[s*i for i in v]
```

```python
#add squres
l = [1,2,3,4,5]
[i**2 for i in l]
```

```python
[i for i in range(1,51) if i % 5 == 0]
```

```python
languages = ['Java', 'C', 'C++', 'Python', 'PHP', 'Javascript']
[lan for lan in languages if lan.startswith('P')]
```
