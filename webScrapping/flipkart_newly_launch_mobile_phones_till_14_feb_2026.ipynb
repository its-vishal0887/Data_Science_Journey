```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from itertools import zip_longest
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
    }
```

```python
names = []
prices = []
rating = []
ram_and_rom = []
display = []
camera = []
processor = []
warrenty = []
image_link = []

for j in range(1, 41):
    link = 'https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&marketplace=FLIPKART&sort=popularity&page={}'.format(j)
    webpage = requests.get(link, headers = headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    soup.prettify()
    for i in range(1, 22):
        phone = soup.find_all('div', class_ = 'jIjQ8S')[i]
        name = phone.find('div', class_ = 'RG5Slk').text.strip()
        price = phone.find('div', class_ = 'hZ3P6w DeU9vF').text.strip()
        rats = soup.find_all('div', class_='a7saXW')[i]
        mats = rats.find('span', class_ = 'PvbNMB').text.split()
        gets = " ".join(mats)
        img = phone.find('div', class_ = 'lWX0_T')
        imgs = str(img).split(" ")
        cng = imgs[-1].split('/>')
        cng[0]
        image_link.append(cng[0])
        rating.append(gets)
        names.append(name)
        prices.append(price)
        for k in range(0,4):
            det =  phone.find_all('li', class_ = 'DTBslk')[k]
            ket = det.text.strip()
            if k == 0:
                ram_and_rom.append(ket)
            elif k == 1:
                display.append(ket)
            elif k == 2:
                camera.append(ket)
            elif k == 3:
                processor.append(ket)
            else:
                warrenty.append(ket)

print(image_link)
print(warrenty)
print(processor)
print(camera)
print(display)
print(ram_and_rom)
print(rating)
print(prices)
print(names)
```
