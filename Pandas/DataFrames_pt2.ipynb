```python
points = pd.DataFrame(
    {
        '1st_points': [(3,4),(-6,5),(0,0),(4,5)],
        '2nd_points':[(-3,4),(0,0),(2,2),(1,1)]
    }
)
```

```python
points
```

```python
def euclidianPts(row):
  pt_A = row['1st_points']
  pt_B = row['2nd_points']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points['euclidianDist'] = points.apply(euclidianPts, axis=1)
```
